# Machine Learning Problem 4


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import progressbar
```

```python
def dotmul(mat1, mat2):
    """
    Dot product multipication
    """
    return np.dot(mat1,mat2)
```

## Problem 4

```python
sicodata = pd.read_csv('siCoData.csv')
sicodata.head(2)
```

```python
# Creating X, Y data
X = pd.concat([ pd.Series([1]*sicodata.shape[0]), sicodata.iloc[:,0]], 1)  
Y = sicodata.iloc[:,1]
```

```python
progress = progressbar.ProgressBar()

"""
Begin Neural Network
"""

# error at every epoch
error_epoch = []

# all errors
error_all = []


# All weights
w1_0_all = []
w1_1_all = []
w2_0_all = []
w2_1_all = []


# initialize weights of layer 1 
W1 = np.array([0.5,0.5])

# weights of the output layer
W2 =  np.array([0.5,0.5])


# Predictions using final weights
Yhat = []

num_epochs = 200

for epoch in progress(range(num_epochs)):
    #print (epoch)
    
    for i in range(sicodata.shape[0]):
                    
        # First Layer ---------------------

            # Get the Z values of layer 1
        Z1 = dotmul(W1.T, X.values[i].T)

            # Apply activation function of layer 1
        A = np.tanh(Z1)
       # dimensions: 1x1 
    
        A1 = np.append(1,A) 

        #Second Layer--------------------

            # Get the Z values of layer 2
        Z2 =  dotmul(W2.T, A1.T)  # Appending the bias to output of layer 1 here

            # Do not apply activation function for the last layer
        A2 = Z2 
        # dimensions: 1x1

        #Error--------------------------    
        #error1 = nn_error(Y[i],A2)
        error = (A2-Y[i])**2
                
        #Backpropogation------------- 

            # Derive Derivatives

        alpha = 0.001
        
        # Update weights----------------------    

        # Update layer 1 weights

        W1[0] = W1[0] - (alpha *((2*(A2-Y[i])) * (W2[1]) * (1- (A1[1]**2)) * X.values[i][0])) # First weight of first layer
        W1[1] = W1[1] - (alpha *((2*(A2-Y[i])) * (W2[1]) * (1- (A1[1]**2)) * X.values[i][1])) # first weight of second layer
        
        # Update layer 2 weights

        W2[0] = W2[0] - (alpha*((2*(A2-Y[i])) * (A1[0])))  # FIRST weight of second layer 
        W2[1] = W2[1] - (alpha*((2*(A2-Y[i])) * (A1[1])))  # Second weight of the second layer
    
    
        # append every error
        error_all.append(error)
                
    # append error at the end of each epoch
    error_epoch.append(error)
    
    # append weights at the end of every epoch
    w1_0_all.append(W1[0]), w1_1_all.append(W1[1]), w2_0_all.append(W2[0]), w2_1_all.append(W2[1])
```

## Plotting

```python
# Plotting the Error and every Epoch and the weights

plt.figure(num=None, figsize=(12, 12), dpi=80, facecolor='w', edgecolor='k')


plt.subplot(2,2,1)
plt.title('Error at every epoch')
plt.plot(error_epoch)

plt.subplot(2,2,2)
plt.title('All weights at each epoch')
for x,y in zip([w1_0_all, w1_1_all, w2_0_all, w2_1_all], ['w10', 'w11', 'w20', 'w21']):
    plt.plot(x, label = '{}'.format(y))
    
plt.legend()
```

```python
# All weights
print(w1_0_all[-1])
print(w1_1_all[-1])
print(w2_0_all[-1])
print(w2_1_all[-1])
```

```python
# prediction

def preds(X):
    
    all_preds = []
    
    for i in range(X.shape[0]):
    

        W1 = np.array([w1_0_all[-1],w1_1_all[-1]]) 
        W2 = np.array([w2_0_all[-1],w2_1_all[-1]]) 

        Z1 = dotmul(W1.T, X.values[i].T)

            # Apply activation function of layer 1
        A = np.tanh(Z1)
        # dimensions: 1x1 

        A1 = np.append(1,A) 

        #Second Layer--------------------

            # Get the Z values of layer 2
        Z2 =  dotmul(W2.T, A1.T)  # Appending the bias to output of layer 1 here

            # Do not apply activation function for the last layer
        all_preds.append(Z2)
    return all_preds
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
# plotting original data
plt.scatter(sicodata.iloc[:,0],sicodata.iloc[:,1] ,label= 'True Values')
plt.legend()
plt.xlabel('X values')
plt.ylabel('True values')

plt.subplot(2,2,2)
# plotting predicted data
plt.scatter(sicodata.iloc[:,0],preds(X), c = 'red', label = 'Predictions')
plt.legend()
plt.xlabel('X values')
plt.ylabel('Predicted values')


plt.figure(figsize=(20,10))
# plotting predicted data
plt.scatter(sicodata.iloc[:,0],sicodata.iloc[:,1] ,label= 'True Values')
plt.scatter(sicodata.iloc[:,0],preds(X), c = 'red', label = 'Predictions')
plt.legend()
plt.xlabel('X values')
plt.ylabel('Predicted values')

```
